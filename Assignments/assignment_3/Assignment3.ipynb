(e) Classifying a Variable

<font size = "3">

- Below, the variable `pizza_rating` is assigned a random value between 0 and 10
- Write an if/elif/else statement that does the following
  - `print("Best Pizza Ever")` if the rating is **greater than or equal to 9**.  
  - `print("Pretty Good")` if the rating is **greater than or equal to 7 but less than 9**.  
  - `print("Meh")` if the rating is **greater than or equal to 4 but less than 7**.  
  - `print("Never Again")` otherwise.

- To get full points you must do all of the above tasks 

```python
pizza_rating = np.random.uniform(low = 0, high = 10)

# Write your answer here

if pizza_rating >= 9:
    print('Best Pizza Ever')
elif 9 > pizza_rating >= 7:
    print('Pretty Good')
elif 7 > pizza_rating >= 4:
    print('Meh')
else:
    print('Never Again')

print(pizza_rating)
```

```python
#10 consecutive times
pizza_rating = np.random.uniform(low = 0, high = 10, size = 10)

for r in pizza_rating:
    if r >= 9:
        print(r, 'Best Pizza Ever')
    elif 9 > r >= 7:
        print(r, 'Pretty Good')
    elif 7 > r >= 4:
        print(r, 'Meh')
    else:
        print(r, 'Never Again')
```
