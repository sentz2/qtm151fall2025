````python
# Write your answer
indonesia_emissions = pd.read_csv('data/indonesia_co2_emissions.csv')
print(type(indonesia_emissions))
indonesia_year = "Year"
indonesia_total_emissions = "total_emissions"
indonesia_list = [indonesia_year, indonesia_total_emissions]
indonesia_list_year = indonesia_list[0]
indonesia_list_total_emissions = indonesia_list[1]
print(indonesia_list_year)
print(indonesia_list_total_emissions)
indonesia_emissions[indonesia_list_total_emissions].describe()
indonesia_emissions[indonesia_list_total_emissions].plot()
plt.title("Indonesia CO2 Emissions")
plt.xlabel("year")
plt.ylabel("total_emissions")
plt.show()
print(indonesia_list)
print(type(indonesia_list))


china_emissions = pd.read_csv('data/china_co2_emissions.csv')
print(type(china_emissions))
china_year = "Year"
china_total_emissions = "total_emissions"
china_list = [china_year, china_total_emissions]
china_list_year = china_list[0]
china_list_total_emissions = china_list[1]
print(china_list_year)
print(china_list_total_emissions)
china_emissions[china_list_total_emissions].describe()
china_emissions[china_list_total_emissions].plot()
plt.title("China CO2 Emissions")
plt.xlabel("year")
plt.ylabel("total_emissions")
plt.show()
````
