```python
#linspace
x = np.linspace(0, 2*np.pi, 100)   # from 0 to 2Ï€, 100 points
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
#For-loop plotting
import matplotlib.pyplot as plt

list_vectors = [y1, y2]
for vec in list_vectors:
    plt.plot(x, vec)

plt.xlabel("x")
plt.ylabel("y")
plt.title("Sine and Cosine")
plt.legend(["sin(x)", "cos(x)"])
plt.show()
```

```python
# ============================================================
# ðŸ“ Exam Notes â€“ Detailed Cheatsheet with Key Comments
# ============================================================

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# ------------------------------------------------------------
# 1. if/elif/else block to test membership in two lists
# ------------------------------------------------------------
fruit = "apple"
list_A = ["apple", "banana"]
list_B = ["orange", "pear"]

# âš ï¸ æ˜“é”™ç‚¹: å¿…é¡»ç”¨ if/elif/elseï¼Œä¸è¦å†™å¤šä¸ª if
if fruit in list_A:
    print(fruit, "belongs to List A")
elif fruit in list_B:
    print(fruit, "belongs to List B")
else:
    print(fruit, "does not belong to either list")

# ------------------------------------------------------------
# 2. From two lists -> NumPy array -> histogram
# ------------------------------------------------------------
x_vals = [1, 2, 3, 4]
y_vals = [2, 4, 6, 8]

# âš ï¸ æ˜“é”™ç‚¹: å¿…é¡»æŠŠ list è½¬æˆ np.array æ‰èƒ½é€å…ƒç´ ç›¸ä¹˜
arr = np.array(x_vals) * np.array(y_vals)

# âš ï¸ æ˜“é”™ç‚¹: é¢˜ç›®å¸¸è¦æ±‚â€œchange the colorâ€ -> color="green"
plt.hist(arr, bins=5, color="green")   
plt.xlabel("Value")    # âš ï¸ ä¸€å®šè¦å†™ axis labels
plt.ylabel("Frequency")
plt.title("Histogram of x*y")  # âš ï¸ Title å¿…é¡»æœ‰
plt.show()

# ------------------------------------------------------------
# 3. Random samples from two distributions -> side-by-side histograms
# ------------------------------------------------------------
# âš ï¸ æ˜“é”™ç‚¹: ç”¨ np.random.normal / np.random.uniform
sample1 = np.random.normal(loc=0, scale=1, size=1000)   # mean=0, std=1
sample2 = np.random.uniform(low=-2, high=2, size=1000)  # uniform range

# âš ï¸ æ˜“é”™ç‚¹: side by side è¦ç”¨ subplots(1,2)
fig, axes = plt.subplots(1, 2, figsize=(10,4))

axes[0].hist(sample1, bins=20, color="blue")
axes[0].set_title("Normal(0,1)")
axes[0].set_xlabel("Value")     # æ¯ä¸ªå­å›¾éƒ½è¦å•ç‹¬ label
axes[0].set_ylabel("Frequency")

axes[1].hist(sample2, bins=20, color="red")
axes[1].set_title("Uniform(-2,2)")
axes[1].set_xlabel("Value")
axes[1].set_ylabel("Frequency")

plt.tight_layout()   # âš ï¸ é¿å…å­å›¾é‡å 
plt.show()

# ------------------------------------------------------------
# 4. Read CSV -> DataFrame -> scatter plot
# ------------------------------------------------------------
df = pd.read_csv("example.csv")   # âš ï¸ å¿…é¡»å†™ pd.read_csv
plt.scatter(df["col_x"], df["col_y"], color="purple")  # âš ï¸ scatter vs plot
plt.xlabel("col_x")   # âš ï¸ è¦å’Œ DataFrame åˆ—åä¸€è‡´
plt.ylabel("col_y")
plt.title("Scatter Plot of col_x vs col_y")
plt.show()

# ------------------------------------------------------------
# 5. Loop over list of columns -> print means
# ------------------------------------------------------------
columns = ["col1", "col2", "col3"]
# âš ï¸ æ˜“é”™ç‚¹: ç”¨ df[col].mean()ï¼Œä¸è¦ np.mean(df[col]) ï¼ˆä¸¤è€…éƒ½è¡Œï¼Œä½†è€å¸ˆå¸¸çœ‹è¿™ä¸ªï¼‰
for col in columns:
    print("Mean of", col, "=", df[col].mean())

# ------------------------------------------------------------
# 6. List of CSVs -> loop -> store means in a list
# ------------------------------------------------------------
filenames = ["file1.csv", "file2.csv", "file3.csv"]
means_list = []

# âš ï¸ æ˜“é”™ç‚¹: for å¾ªçŽ¯é‡Œå¿…é¡» appendï¼Œä¸è¦ç›´æŽ¥è¦†ç›–
for fname in filenames:
    temp_df = pd.read_csv(fname)
    mean_val = temp_df["target_col"].mean()  # âš ï¸ target_col è¦æ¢æˆé¢˜ç›®æŒ‡å®šçš„åˆ—
    means_list.append(mean_val)

print("Means from all files:", means_list)
```
